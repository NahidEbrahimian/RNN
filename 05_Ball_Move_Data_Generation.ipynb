```python
# Imports

import numpy as np
import random
import cv2
from google.colab.patches import cv2_imshow
```

```python
dataset_size = 10
num_frames = 45
frame_width = 90
frame_height = 90
step = 2

labels = []
data = []
dataset = []
```

```python
def make_frame(label, init_x, init_y):
  frames = []
  next_y = init_y
  next_x = init_x

  for i in range(num_frames):

    frame = np.zeros((frame_width, frame_height))
    cv2.circle(frame, (next_y, next_x), 5, 255, -1) 

    next_x = random.randint(next_x - 2, next_x + 2)

    if label==1:
      next_y += step 

    else:
      next_y -= step

    frames.append(frame)

  return frames
```

```python
def generate_data(initial_state_x, random_num):

  if random_num < 0.5:
    label = 0
    initial_state_y = frame_height - 1

  else:
    label = 1
    initial_state_y = 1

  frames = make_frame(label, initial_state_x, initial_state_y)

  return frames, label
```

```python
for d in range(dataset_size):
  random_num = random.random()
  initial_state_x = random.randint(2, frame_height - 2)

  data, label = generate_data(initial_state_x, random_num)

  dataset.append(data)
  labels.append(label)
```

```python
# Convert to np.array

dataset = np.array(dataset)
labels = np.array(labels)

dataset.shape, labels.shape
```

```python
# Reshape dataset

dataset = dataset[..., np.newaxis]
labels = labels[..., np.newaxis]

dataset.shape, labels.shape
```

```python
# Plot

for frame in range(len(dataset[8, :])):
    print("frame:", frame+1)
    cv2_imshow(dataset[0,frame, :, :])
    print("\n")
```
